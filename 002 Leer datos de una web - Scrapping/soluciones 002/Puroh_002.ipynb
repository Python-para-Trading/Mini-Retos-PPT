# Beautifullsop4

```python
from requests import get
from bs4 import BeautifulSoup
```

```python
def get_html(url):
    
    # Capturamos el html de la pagina web y creamos un objeto Response
    r  = get(url)
    data = r.text
    
    #si el request tuvo exito
    status_code = r.status_code
    if status_code == 200:        
        # Creamos el objeto soup y le pasamos lo capturado con request
        soup = BeautifulSoup(data, 'lxml')
 
        # Capturamos el titulo de la página y luego lo mostramos
        # Lo que hace BeautifulSoup es capturar lo que esta dentro de la etiqueta title de la url
        
        
    else:
        print ("Status Code %d" % status_code)
    
    
    return soup

def get_data_title(soup):
    head_table = soup.find_all('table', {'id': 'tableOpcionesAcciones'})
    columns_head = head_table[0].find_all('th')
    list_head_titles = []
    for column in columns_head:    
        list_head_titles.append(column.getText().replace('\xa0',''))
    return list_head_titles
    
```

```python
soup = get_html(url)
head_table = soup.find_all('table', {'id': 'tableOpcionesAcciones'})
columns_head = head_table[0].find_all('th')
list_head_titles = []
for column in columns_head:    
    list_head_titles.append(column.getText().replace('\xa0',''))
    
print(list_head_titles)
```

```python
body_table = soup.find_all('table', {'id': 'tableOpcionesAcciones'})
columns_body = body_table[0].find_all('td')
list_body_titles = []
for column in columns_body[:14]:    
    list_body_titles.append(column.getText().replace('\xa0','').replace(' ','').replace('\n','').replace('\r',''))
print(list_body_titles)
```

```python
len(columns_body)/15 #Para verificar si los datos estan completos 
```

```python
list_body = []
list_body_titles = []
preslice = 0
for slices in range(15, len(columns_body),15):
    for column in columns_body[preslice:slices]: #cada fila de datos va de 15 en 15 entonces se toma cada 15 valores para completar la fila y pasarla al siguiente for
        list_body_titles.append(column.getText().replace('\xa0','').replace('\n','').replace('\r','').replace('Vermás', ''))#se eliminan saltos de linea y demas caracteres especiales, y se toman los datos en una lista
    list_body_titles = []
    list_body.append(list_body_titles)#luego cada fila se agrega a otra lista, queda una lista de listas [[fila1],[fila2],[fila3], ...]
    preslice = slices
```

```python
list_body[:2][:14]# muestro los datos de como quedan guardados
```

Se toman los primeros 15 datos y se descartan el resto

```python
list_datos = []
for slices in range(0, len(list_body)):
    list_datos.append(list_body[slices][:14])
list_datos
```

Se eliminan las filas con valores None

```python
df=pd.DataFrame(list_datos, columns=list_head_titles).dropna(how='all')
df
```

```python
df = df.set_index('Especie') # se cooloca la columna 'Especie' como indice del dataframe
```

Se cambia el formato por formatoo Datetime

```python
df[['Vencimiento', 'Hora']] = df[['Vencimiento', 'Hora']].astype('datetime64[ns]')
```

```python
df[['Cant. venta','Precio ejercicio', 'Último precio (prima)', 'Precio compra', 'Precio venta', 'Cant. venta', 'Volumen (lotes)', 'Volumen (monto)', 'Cant. compra' ]] = df[[ 'Cant. venta','Precio ejercicio', 'Último precio (prima)', 'Precio compra', 'Precio venta', 'Cant. venta', 'Volumen (lotes)', 'Volumen (monto)', 'Cant. compra']].applymap(lambda x: x.replace('.','').replace(',','.') if x is not None else x).astype('float64')
df.dtypes
```
