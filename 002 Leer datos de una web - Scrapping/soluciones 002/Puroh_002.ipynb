Pandas toma todos los datos de la pagina que se esta haciendo scrapping y los organiza en un lista de dataframes, es decir [dataframe, dataframe2, ... , dataframeN], como en la pagina que se le hace el scrapping solo tiene una tabla, entonces solo nos interesa los valores que tiene el primer elemento de la lista, por ello se utiliza df_list[0]

ahora se ahce un iloc, con este se accede a un grupo de filas o columnas utilizando enteros para la posicion o posiciones. Para las filas se utiliza una función lambda que valida el indice numerico y seleciona solo los pares, que son las filas qe contienen la información solicitada, y el segundo slice de las columnas se coloca 14 para las primeras catorce columnas de la tabla.

Con .set_index('Especie') se selecciona la columna que se desea 

el "inplace = True" es para que se modifique automaticamente el dataframe con el que se esta trabajando 

```python
df=df_list[0].iloc[lambda x: x.index % 2 == 0, :14].set_index('Especie')#,inplace = True)
df
```

# Beautifullsop4

```python
from requests import get
from bs4 import BeautifulSoup
```

```python
def get_html(url):
    
    # Capturamos el hml de la pagina web y creamos un objeto Response
    r  = get(url)
    data = r.text
    
    #si el request tuvo exito
    status_code = r.status_code
    if status_code == 200:        
        # Creamos el objeto soup y le pasamos lo capturado con request
        soup = BeautifulSoup(data, 'lxml')
 
        # Capturamos el titulo de la página y luego lo mostramos
        # Lo que hace BeautifulSoup es capturar lo que esta dentro de la etiqueta title de la url
        
        
    else:
        print ("Status Code %d" % status_code)
    
    
    return soup

def get_data_title(soup):
    head_table = soup.find_all('table', {'id': 'tableOpcionesAcciones'})
    columns_head = head_table[0].find_all('th')
    list_head_titles = []
    for column in columns_head:    
        list_head_titles.append(column.getText().replace('\xa0',''))
    return list_head_titles
    
```

```python
soup = get_html(url)
head_table = soup.find_all('table', {'id': 'tableOpcionesAcciones'})
columns_head = head_table[0].find_all('th')
list_head_titles = []
for column in columns_head:    
    list_head_titles.append(column.getText().replace('\xa0',''))
    
print(list_head_titles)
```

```python
body_table = soup.find_all('table', {'id': 'tableOpcionesAcciones'})
columns_body = body_table[0].find_all('td')
print(columns_body[13].getText())
list_body_titles = []
for column in columns_body[:14]:    
    list_body_titles.append(column.getText().replace('\xa0','').replace(' ','').replace('\n','').replace('\r',''))
print(list_body_titles)
```
