## Paduel @paduel

Utilizar Series en df2 en lugar de DataFrame. De forma general para Pandas, no useis dataframe cuando realmente solo teneis una columna, usad Series, a menos que haya un buen motivo para ello.

Es un caso interesante, pues no se puede usar ni reindex ni rename, pues la repitición de columnas haría que eliminara esas duplicadas.

Nota:
.squeeze() pasa un dataframe a serie si es posible
.to_frame(col) pasa una serie a dataframe, con el nombre de columna que se pase como parametro col

```python
df2 = pd.Series({'A': 'SS1',
                 'B': 'SS3',
                 'C': 'SS4',
                 'D': 'SS7',
                 'E': 'SS1',
                 'F': 'SS6',
                 'G': 'SS5',
                 'H': 'SS1',
                 'I': 'SS5',
                 'J': 'SS2'})

df3 = df1[df2]
df3.columns = df2.index
df3
```
